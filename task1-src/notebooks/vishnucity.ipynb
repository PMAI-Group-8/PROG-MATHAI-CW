```python
from Cifar10Loader import Cifar10Loader
from neural_network import NeuralNetwork
from neuron_layer import NeuronLayer

import numpy as np
np.random.seed(42)
```

```python
CIFAR_DIR = "..\dataset\cifar10"  # adjust if needed

loader = Cifar10Loader(CIFAR_DIR)

X_train_raw, y_train = loader.load_train_data()
X_test_raw, y_test = loader.load_test_data()

print(X_train_raw.shape, y_train.shape)
```

```python
def preprocess(X, y, shuffle=True):
    # normalise
    X = X.astype(np.float32) / 255.0

    # reshape to image format (optional but correct)
    X = X.reshape(-1, 3, 32, 32).transpose(0, 2, 3, 1)

    # flatten for MLP
    X = X.reshape(X.shape[0], -1)

    # shuffle
    if shuffle:
        idx = np.random.permutation(X.shape[0])
        X, y = X[idx], y[idx]

    return X, y
```

```python
X_train, y_train = preprocess(X_train_raw, y_train)
X_test, y_test = preprocess(X_test_raw, y_test, shuffle=False)

print(X_train.shape, y_train.shape)
```

```python
nn = NeuralNetwork(learning_rate=0.01)
nn.train(X_train, y_train, epochs=100)
```

```python
logits = nn.forward(X_test)
preds = np.argmax(logits, axis=1)
acc = np.mean(preds == y_test)
print("Test accuracy:", acc)
```
